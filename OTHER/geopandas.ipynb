```python
import pandas as pd
import folium
import numpy as np
import branca.colormap as cm

# Create a sample DataFrame
np.random.seed(42)  # for reproducibility

data = {
    'locationName': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 
                     'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'latitude': [40.7128, 34.0522, 41.8781, 29.7604, 33.4484, 
                 39.9526, 29.4241, 32.7157, 32.7767, 37.3382],
    'longitude': [-74.0060, -118.2437, -87.6298, -95.3698, -112.0740, 
                  -75.1652, -98.4936, -117.1611, -96.7970, -121.8863],
    'intensity': np.random.randint(5, 20, 10)  # Random values between 5 and 20
}

df = pd.DataFrame(data)

print("Sample DataFrame:")
# print(df)

# Create a map centered on the mean latitude and longitude
center_lat = df['latitude'].mean()
center_lon = df['longitude'].mean()
map = folium.Map(location=[center_lat, center_lon], zoom_start=4)

# Create a colormap
colormap = cm.LinearColormap(colors=['blue', 'yellow', 'red'], vmin=df['intensity'].min(), vmax=df['intensity'].max())
map.add_child(colormap)

# Add bubbles for each location
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=row['intensity'],
        popup=row['locationName'],
        tooltip=f"Location: {row['locationName']}<br>Intensity: {row['intensity']}",
        color=colormap(row['intensity']),
        fill=True,
        fill_color=colormap(row['intensity']),
        fill_opacity=0.7
    ).add_to(map)

# Save the map
# map.save('interactive_intensity_bubble_map.html')

print("\nMap has been saved as 'interactive_intensity_bubble_map.html'")

# If you're in a Jupyter notebook, you can display the map with:
display(map)
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import numpy as np

# Create a sample DataFrame
np.random.seed(42)  # for reproducibility

data = {
    'locationName': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 
                     'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'latitude': [40.7128, 34.0522, 41.8781, 29.7604, 33.4484, 
                 39.9526, 29.4241, 32.7157, 32.7767, 37.3382],
    'longitude': [-74.0060, -118.2437, -87.6298, -95.3698, -112.0740, 
                  -75.1652, -98.4936, -117.1611, -96.7970, -121.8863],
    'intensity': np.random.randint(5, 20, 10)  # Random values between 5 and 20
}

df = pd.DataFrame(data)

print("Sample DataFrame:")
# print(df)

# Create a map centered on the mean latitude and longitude
center_lat = df['latitude'].mean()
center_lon = df['longitude'].mean()
map = folium.Map(location=[center_lat, center_lon], zoom_start=4)

# Prepare the data for the heatmap
heat_data = [[row['latitude'], row['longitude'], row['intensity']] for index, row in df.iterrows()]

# Add the heatmap to the map
HeatMap(heat_data, radius=50, blur=10, max_zoom=1, overlay=True).add_to(map)

# Add markers for each location with popups
for idx, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"Location: {row['locationName']}<br>Intensity: {row['intensity']}",
        tooltip=row['locationName']
    ).add_to(map)

# Save the map
# map.save('intensity_heatmap.html')

print("\nMap has been saved as 'intensity_heatmap.html'")

# If you're in a Jupyter notebook, you can display the map with:
# display(map)
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import numpy as np
import branca.colormap as cm

# Create a sample DataFrame
np.random.seed(42)  # for reproducibility

data = {
    'locationName': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 
                     'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'latitude': [40.7128, 34.0522, 41.8781, 29.7604, 33.4484, 
                 39.9526, 29.4241, 32.7157, 32.7767, 37.3382],
    'longitude': [-74.0060, -118.2437, -87.6298, -95.3698, -112.0740, 
                  -75.1652, -98.4936, -117.1611, -96.7970, -121.8863],
    'intensity': np.random.randint(5, 20, 10)  # Random values between 5 and 20
}

df = pd.DataFrame(data)

print("Sample DataFrame:")
# print(df)

# Create a map centered on the mean latitude and longitude
center_lat = df['latitude'].mean()
center_lon = df['longitude'].mean()
map = folium.Map(location=[center_lat, center_lon], zoom_start=4)

# Prepare data for heatmap
heat_data = [[row['latitude'], row['longitude'], row['intensity']] for index, row in df.iterrows()]

# Add heatmap layer
HeatMap(heat_data).add_to(map)

# Create a colormap for the bubbles
colormap = cm.LinearColormap(colors=['blue', 'yellow', 'red'], vmin=df['intensity'].min(), vmax=df['intensity'].max())
map.add_child(colormap)

# Add bubbles for each location
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=row['intensity'],
        popup=row['locationName'],
        tooltip=f"Location: {row['locationName']}<br>Intensity: {row['intensity']}",
        color=colormap(row['intensity']),
        fill=True,
        fill_color=colormap(row['intensity']),
        fill_opacity=0.7,
        weight=2
    ).add_to(map)

# Add a layer control
folium.LayerControl().add_to(map)

# Save the map
# map.save('heatmap_with_bubbles.html')

print("\nMap has been saved as 'heatmap_with_bubbles.html'")

# If you're in a Jupyter notebook, you can display the map with:
display(map)
```

```python
import folium
import pandas as pd
import geopandas as gpd

# Sample DataFrame
data = {
    'State': ['California', 'Texas', 'Florida', 'New York', 'Pennsylvania'],
    'Population': [39538223, 29145505, 21538187, 20201249, 13002700]
}
df = pd.DataFrame(data)

# GeoJSON file path
geo_data = 'C:\\Users\\darsh\\vscodeProjects\\vscodeProjectQuickTest\\nursingProject\\Data\\us-states.json'

# Read GeoJSON file
gdf = gpd.read_file(geo_data)

# Merge GeoDataFrame with population data
gdf = gdf.merge(df, left_on='NAME', right_on='State', how='inner')

# Create map
m = folium.Map(location=[37.8, -96], zoom_start=4)

# Add choropleth layer
choropleth = folium.Choropleth(
    geo_data=gdf,
    name='choropleth',
    data=df,
    columns=['State', 'Population'],
    key_on='feature.properties.NAME',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Population'
).add_to(m)

# Add hover functionality
style_function = lambda x: {'fillColor': '#ffffff', 
                            'color':'#000000', 
                            'fillOpacity': 0.1, 
                            'weight': 0.1}
highlight_function = lambda x: {'fillColor': '#000000', 
                                'color':'#000000', 
                                'fillOpacity': 0.50, 
                                'weight': 0.1}

# Add GeoJson layer with tooltip
folium.GeoJson(
    gdf,
    style_function=style_function,
    control=False,
    highlight_function=highlight_function,
    tooltip=folium.features.GeoJsonTooltip(
        fields=['STATE', 'Population'],
        aliases=['State: ', 'Population: '],
        style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;"),
        localize=True,
        sticky=False
    )
).add_to(m)

# Save map
# m.save('us_states_population.html')

# Display map (this will work in a Jupyter notebook)
display(m)
```
